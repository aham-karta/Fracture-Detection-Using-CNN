```python
import matplotlib.pyplot as plt
import numpy as np

# Data
values1 = [61.292,
65.795,
66.047,
94.72,
97.673,
95.415,
97.97,
99.01,
97.85,
98.59,
98.68,
98.5,
95.56,
98.39,
98.5]
values2 = [75.9382,
75.3452,
71.4563,
94.6078,
98.4694,
97.1722,
98.7654,
98.7147,
99.4949,
99.0781,
99.4979,
99.4737,
99.7658,
99.4924,
99.7602]
values1.reverse()
values2.reverse()

# Ensure the data lists have the same length as categories
categories = [f'file{i}' for i in range(len(values1), 0, -1)]

# Create a figure and axis
fig, ax = plt.subplots()

# Set bar width
bar_width = 0.35

# Set the position of the bars on the y-axis
y_pos = np.arange(len(categories))

# Create bars (plot red bars on top)
bars1 = ax.barh(y_pos + bar_width/2, values1, bar_width, label='Average of 10 runs', color='lightcoral')
bars2 = ax.barh(y_pos - bar_width/2, values2, bar_width, label='K Fold Cross Validation (10)', color='lightskyblue')

# Add labels, title, and legend
ax.set_yticks(y_pos)
ax.set_yticklabels(categories)
ax.set_xlabel('F1 Score')
ax.set_title('Comparative Horizontal Bar Chart')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_minor_locator(plt.MultipleLocator(1))
ax.set_xlim(0, 100)  # Set the x-axis range from 0 to 100
ax.xaxis.set_major_locator(plt.MultipleLocator(10))  # Set major ticks every 10 units
ax.xaxis.set_minor_locator(plt.MultipleLocator(1)) 

# Move the legend below the graph
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=1)

# Display the plot
plt.show()
```
