```python
import matplotlib.pyplot as plt
import numpy as np

values = [
    72.9566, 69.2656, 65.0967, 94.6243, 98.5131, 
    97.4537, 98.7756, 98.7635, 99.4975, 99.0183, 
    99.5105, 99.4764, 99.7664, 99.5305, 99.7512
]
labels = [f'file{i}' for i in range(1, 16)]

fig, ax = plt.subplots()
ax.plot(labels, values, label='Area Under Curve', color='mediumpurple', marker='o',markersize=2)

ax.set_xlabel('Files')
ax.set_ylabel('AUC')
ax.set_title('Comparative Line Chart (AUC)')
ax.set_ylim(60, 110) 
ax.yaxis.set_minor_locator(plt.MultipleLocator(1))  

ax.legend()

plt.xticks(rotation=45)
plt.show()
```
